```
import random
from random import randint
board = [0, 1, 2,
         3, 4, 5,
         6, 7, 8]
type(board) #type of board
#function does take any arugument simple draw a baord
def tictactoe_draw():
  print(board[0], board[1], board[2])
  print(board[3], board[4], board[5])
  print(board[6], board[7], board[8]) 

#Player 1 input
def Player_1():
  print("Player 1 : Enter your desired location : ")
  location = int(input())
  if board[location] !='X' or board[location] != 'O':  #empty spot
    board[location] = 'X'
  else:
    print("The spot is taken place. Please choose another")
tictactoe_draw()

Player_1()
tictactoe_draw() # Draw the table

#Computer (Player 2) input
def Player_2():
  print("Player 2 : Enter your desired location : ")
  location = randint(0, 9)
  if board[location] !='X' and board[location] != 'O':
    board[location] = 'O'
  else:
    print("The spot is taken place. Please choose another")
   
tictactoe_draw()
Player_2()
tictactoe_draw()

#Check if any player wins!
def Check_Winner(X_or_O):
  if board[0] == X_or_O and board[1] == X_or_O and board[2] == X_or_O:
    return True
  elif board[3] == X_or_O and board[4] == X_or_O and board[5] == X_or_O:
    return True
  elif board[6] == X_or_O and board[7] == X_or_O and board[8] == X_or_O:
    return True
  elif board[0] == X_or_O and board[3] == X_or_O and board[6] == X_or_O:
    return True
  elif board[1] == X_or_O and board[4] == X_or_O and board[7] == X_or_O:
    return True
  elif board[2] == X_or_O and board[5] == X_or_O and board[8] == X_or_O:
    return True
  
  #Diagonal check
  if board[0] == X_or_O and board[4] == X_or_O and board[8] == X_or_O:
    return True
  elif board[2] == X_or_O and board[4] == X_or_O and board[6] == X_or_O:
    return True
  
#Let's put it all together
while True:
    Player_1() # Player 1 plays
    tictactoe_draw()
    if Check_Winner ('X') == True:
      print("Player 1 Wins")
      break;
    Player_2()
    tictactoe_draw()
    if Check_Winner ('O') == True:
      print("Player 2 Wins")
      break;

print("Thank you for Playing Tic Tac Toe Game")



```
